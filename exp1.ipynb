```python
import ollama
import re

model_name = "codellama:7b"  # change if needed

prompt = """
Write three Python functions for calculating the factorial of a non-negative integer.
Each function should include input validation to handle negative numbers or non-integer inputs by raising a ValueError.

1.  factorial_iterative: Calculates factorial using a loop (iterative approach).
2.  factorial_recursive: Calculates factorial using recursion.
3.  factorial_builtin_or_functional: Calculates factorial using Python's built-in math.factorial or a functional approach.
"""

response = ollama.generate(
    model=model_name,
    prompt=prompt
)

generated_code_raw = response["response"]
print("Generated Code from Ollama Model (raw output):")
print(generated_code_raw)

generated_functions = {}
current_function_name = None
current_function_code = []
lines = generated_code_raw.split('\n')

pattern = r"def\s+(factorial_(iterative|recursive|builtin_or_functional))\s*\(.*\):"

for line in lines:
    match = re.match(pattern, line.strip())
    if match:
        if current_function_name:
            generated_functions[current_function_name] = "\n".join(current_function_code).strip()
        current_function_name = match.group(1)
        current_function_code = [line]
    elif current_function_name:
        current_function_code.append(line)

if current_function_name:
    generated_functions[current_function_name] = "\n".join(current_function_code).strip()

print("\n--- Separated Generated Factorial Functions ---")
for name, code in generated_functions.items():
    print(f"\nFunction: {name}")
    print(code)
    print("-" * 40)
```
